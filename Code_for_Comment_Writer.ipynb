```python
import csv
def grade_comments(student):
    grade=int(student[2])
    if grade > 92:
        return "This is an outstanding accomplishment, and I am proud of the hard work and dedication you have shown throughout the semester. Your grade demonstrates your mastery of the course material, and I have no doubt that you will continue to excel in your academic pursuits."
    if grade > 89:
        return "This is a great achievement, and I am impressed with the hard work and dedication you have demonstrated throughout the semester. Your grade reflects your commitment to learning, and I am confident that you will continue to excel in your academic pursuits."
    if grade > 86:
        return "This is a solid performance, and I am proud of the effort you have put in throughout the semester. Your grade demonstrates your understanding of the course material and your dedication to your academic goals."
    if grade > 79:
        return "While this may not be the grade you were hoping for, it is still a solid performance, and I commend you for your hard work and dedication throughout the semester."
    if grade > 73:
        return "While this may not be the grade you were hoping for, I encourage you to take this as an opportunity to reflect on your performance and to identify areas where you can improve. Your grade reflects your current understanding of the course material, and I believe that with some extra effort, you can achieve better results in the future."
    else:
        return "This is a concerning result, and I understand that you may be feeling disappointed or discouraged at this time. However, I encourage you to take this as an opportunity to reflect on your performance and to identify areas where you can improve."


def indicatorcomments(studentinfo):
    problemsolving1=["Identifies multiple notations associated with the derivative of a function."," Demonstrating mastery of using limits to define continuity and evaluating limit expressions using l'Hopital's rule."]
    precision2=["Uses strong foundational skills to go about computations methodically and systematically, especially when computing derivatives.","Written communication is clear and organized vertically in a step-by-step manner."]
    propertool3=["Demonstrating mastery of the algebra involved in computing derivative rules but sometimes gets stuck when trying to create a graphical representation on the calculator.","Looking forward, I encourage you to speak up with questions when you notice that you can not find special features of graphs on the calculator (intersections, max/mins, intercepts, and derivatives)."]
    viablearguments4=["On modeling problems that require an explanation beyond the numerical answer, make sure to include units and provide reasoning in context.","Developing great habits when it comes to checking work for mistakes. Now, consider speaking up when you notice that you've used a different method during group work. Even if you come to the same answer, it would benefit everyone to talk through the efficiency of different methods. For example, what connections can you draw between your process for solving related rates problems and the alternate method of differentiating with respect to time?"]
    indicators = [problemsolving1, precision2, propertool3, viablearguments4]
    good,bad=[],[]
    for n, indicator in enumerate(indicators):
        if int(studentinfo[n+3])>2:
            for i in indicator:
                good.append(i)
        else:
            for i in indicator:
                bad.append(i)

    if bad==[]:
        bad.append("You're totally capable and ready for our next section of studying")
    return ["Strengths:\n",good,"\n","Area for Growth:\n",bad]

def attribute_comment(attribute):
    attribute=attribute[7]
    attribute_dict ={"perserverance": "I commend your persistence and determination in the face of challenges. It's clear that you have been working hard to overcome obstacles and make progress in your studies, and I am proud of the effort you have put in.", "growth": "It is clear that you have been working hard to develop new skills and improve in your academic pursuits, and I am impressed by the strides you have made.","posotive": "Your energy and optimism are truly infectious, and they have a profound impact on those around you.","motivated": "Your eagerness to learn and your commitment to your goals are truly inspiring, and they set you apart from your peers."}
    return attribute_dict[attribute]

def comments():
    with open('sheet1.csv', newline='') as csvfile:
        all = csv.reader(csvfile)
        title=next(all)
        for student in all:
            with open(f'{student[1]}.txt', 'w') as z:
                z.write(f"{student[0]} {student[1]}\n")
                for i in indicatorcomments(student):
                    if type(i)==list:
                        for a in i:
                            z.write(a)
                    else:
                        z.write(i)
                z.write(f"\nPersonal c{attribute_comment(student)}")
                z.write(f"\nGrade comments:\n{grade_comments(student)}\n")

comments()
```
